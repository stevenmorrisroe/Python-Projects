```python
import requests
from bs4 import BeautifulSoup
count = 420
Athlete_Name = []
while count >300:
    url = "https://nilcollegeathletes.com/athletes?page="
    url+=str(count)
    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'html.parser')

    links = soup.find_all('a', class_="hover:underline hover:text-blue-700")

    for link in links:
        name = link.text
        Athlete_Name.append(name)
    count-=1
print(Athlete_Name)
```
